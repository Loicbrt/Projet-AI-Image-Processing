```python
# Le dataset COCO doit être télécharger à la main sur le site de COCO : https://cocodataset.org/#download
# Training dataset : http://images.cocodataset.org/zips/train2014.zip
# Validation dataset : http://images.cocodataset.org/zips/val2014.zip
# Test dataset : http://images.cocodataset.org/zips/test2014.zip
# COCO 2014 est composé d'image de taille varié, d'environ 800x800, une taille bien plus importante que pour MNIST ou CIFAR
# Les images peuvent aussi être en RGB et en noir et blanc

# On convertit les images RGB en un tuple de tensor Lab ([1, 256, 256],[2, 256, 256]) utilisable par pytorch (channels x width x height).
# On change la taille des images en 256x256 pour l'entraînement car toute les images d'un batch doivent être de la même taille. Mais lors de l'utilisation du modèle, on envoie une image normale au modèle
# On crée ensuite deux objets Dataset, l'un pour l'entraînement, l'autre pour la validation. Chaque élément est composé de l'image en niveau de gris (L) et de la couleur (ab)


transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((256, 256)),
])
train_data = ImageFolderDataset(root_dir="data/COCO/train2014", transform=transform)
val_data = ImageFolderDataset(root_dir="data/COCO/val2014", transform=transform)

print("Number of training samples:", len(train_data))
print("Number of validation samples:", len(val_data))
print("Format des images d'entrainement: ("+ str(train_data[0][0].shape)+","+str(train_data[0][1].shape)+")")

show_data(train_data)
```

```python
#Bloc pour lancer l'entraînement d'un autoencoder, voir U_Net et train_model() dans src.model
#Le modèle et les donées d'entraînement sont ensuite sauvegardé à la fin
#L'entraînement est long !!! (environ 12h pour une epoch sur un cypc)

trained_U_net_model, data = train_model(
    train_dataset=(train_data),
    val_dataset=(val_data),
    train_data_file_name = 'U_Net.csv'
)

torch.save(trained_U_net_model.state_dict(), "model/U_Net_model.pth")
```
